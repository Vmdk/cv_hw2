```python
import numpy as np
import cv2
from skimage import io
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow

def url_to_image(url):
  print("downloading %s" % (url))
  return cv2.cvtColor(io.imread(url), cv2.COLOR_BGR2RGB)
```

### Task 1


```python
url_11 = "https://raw.githubusercontent.com/mdavydov/ComputerVisionCourse/master/tasks/text/text1.jpg"
img = url_to_image(url_11)
img = cv2.resize(img[110:640,:], (0,0), fx=0.7, fy=0.7) #960 x 720
grayscaled = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
threshold = cv2.adaptiveThreshold(grayscaled, 255,
                          cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                          cv2.THRESH_BINARY, 15, 10)
cv2_imshow(grayscaled)
cv2_imshow(threshold)
# img.shape
```

```python
url_13 = "https://raw.githubusercontent.com/mdavydov/ComputerVisionCourse/master/tasks/text/text3.jpg"
img = url_to_image(url_13)
img = cv2.resize(img, (0,0), fx=0.7, fy=0.7) 
msf_img = cv2.pyrMeanShiftFiltering(img, 11, 21)
grayscaled = cv2.cvtColor(msf_img,cv2.COLOR_BGR2GRAY)
threshold = cv2.bitwise_not(grayscaled)
rev, threshold = cv2.threshold(threshold, 10, 255, cv2.THRESH_TOZERO)
threshold = cv2.adaptiveThreshold(threshold, 255,
                          cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                          cv2.THRESH_BINARY, 11, 25)
cv2_imshow(img)
cv2_imshow(threshold)
```
