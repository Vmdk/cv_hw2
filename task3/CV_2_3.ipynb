```python
import cv2
import numpy as np
from skimage import io
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow
from skimage.feature import peak_local_max
from skimage.morphology import watershed
from scipy import ndimage

def url_to_image(url):
  return cv2.cvtColor(io.imread(url), cv2.COLOR_BGR2RGB)
```

### Task 3


```python
url31 = "https://raw.githubusercontent.com/mdavydov/ComputerVisionCourse/master/tasks/object/obj1.jpg"

img = url_to_image(url31)
img = cv2.resize(img, (0,0), fx=0.5, fy=0.5) 

cv2_imshow(img)
(h, w) = img.shape[:2]# 134 x 240
mask = np.zeros( (h, w),np.uint8)
bgdModel = np.zeros((1,65),np.float64)
fgdModel = np.zeros((1,65),np.float64)
rect = (10,10,w-25,h-20)
cv2.grabCut(img,mask,rect,bgdModel,fgdModel,2,cv2.GC_INIT_WITH_RECT)
mask2 = np.where((mask==2)|(mask==0),0,1).astype('uint8')
img = img*mask2[:,:,np.newaxis]
cv2_imshow(mask2*255)
cv2_imshow(img)
```
