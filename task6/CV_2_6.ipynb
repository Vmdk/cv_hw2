```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
from skimage import io
from matplotlib import pyplot as plt
import time

def url_to_image(url):
  print("downloading %s" % (url))
  return cv2.cvtColor(io.imread(url), cv2.COLOR_BGR2RGB)
```

```python
class PatchMatch:
  def __init__(self, img_A, img_B, iterations, patchSize):
    print("Initialization...")
    self.A = img_A
    self.A_size = img_A.shape[:2]
    self.B = img_B
    self.B_size = img_B.shape[:2]
    self.iterationsCount = iterations
    self.patchSize = patchSize
    self.halfPadding = patchSize // 2
    
    (A_h, A_w) = self.A_size
    (B_h, B_w) = self.B_size
    h_p = self.halfPadding
    random_B_r = np.random.randint(h_p, B_h-h_p, self.A_size)
    random_B_c = np.random.randint(h_p, B_w-h_p, self.A_size)
    self.imagePadding = np.ones([A_h+h_p*2, A_w+h_p*2, 3]) * np.nan
    self.imagePadding[h_p:A_h+h_p, h_p:A_w+h_p, :] = img_A
    self.f = np.zeros(self.A_size, dtype=object)
    self.dist = np.zeros([A_h, A_w])
    for i in range(A_h):
      for j in range(A_w):
        a = np.array([i, j])
        b = np.array([random_B_r[i, j], random_B_c[i, j]], dtype=np.int32)
        self.f[i, j] = b
        self.dist[i, j] = self.getDistance(a, b)
    
  def getDistance(self,a, b):
    h_p = self.halfPadding
    patch_a = self.imagePadding[a[0]:a[0]+self.patchSize, a[1]:a[1]+self.patchSize, :]
    patch_b = self.B[b[0]-h_p:b[0]+h_p+1, b[1]-h_p:b[1]+h_p+1, :]
    patch_diff = patch_b - patch_a
    pixels_amount = np.sum(1 - np.int32(np.isnan(patch_diff)))
    dist = np.sum(np.square(np.nan_to_num(patch_diff))) / pixels_amount
    return dist
        
        
  def randomSearch(self, x, y):
    alpha=0.5
    (B_h, B_w) = self.B_size
    h_p = self.halfPadding
    i = 4
    search_h = B_h * alpha ** i
    search_w = B_w * alpha ** i
    b_x = self.f[x, y][0]
    b_y = self.f[x, y][1]
    while search_h > 1 and search_w > 1:
      search_min_r = max(b_x - search_h, h_p)
      search_max_r = min(b_x + search_h, B_h-h_p)
      random_b_x = np.random.randint(search_min_r, search_max_r)
      search_min_c = max(b_y - search_w, h_p)
      search_max_c = min(b_y + search_w, B_w - h_p)
      random_b_y = np.random.randint(search_min_c, search_max_c)
      search_h = B_h * alpha ** i
      search_w = B_w * alpha ** i
      b = np.array([random_b_x, random_b_y])
      d = self.getDistance([x,y], b)
      if d < self.dist[x, y]:
        self.dist[x, y] = d
        self.f[x, y] = b
      i += 1
      
      
  def findBestValues(self, coords):
    #first coordinates must be of current point!
    curr_coords = coords[0]
    (x, y) = curr_coords
    #minor mapper from coords to dist
    distances = self.dist[coords[:,0], coords[:,1]]
    bestIndex = np.argmin(distances)
    #set best values, if found one (if coords changed)
    if bestIndex > 0:
      (best_x, best_y) = coords[bestIndex]
      self.f[x, y] = self.f[best_x, best_y]
      self.dist[x, y] = self.getDistance(curr_coords, self.f[x, y])
        
  def oddPropagation(self, x, y):
    curr_coords = np.array([x,y])
    left_x = np.maximum(x-1, 0) # to avoid tryings to take (-1,y) pixels
    left_coords = np.array([left_x, y])
    top_y = np.maximum(y-1, 0)
    top_coords = np.array([x, top_y])
    coords = np.array([curr_coords, left_coords, top_coords])
    self.findBestValues(coords)
  
  def evenPropagation(self, x, y):
    A_h = np.size(self.imagePadding, 0) - self.patchSize + 1
    A_w = np.size(self.imagePadding, 1) - self.patchSize + 1
    right_x = np.minimum(x + 1, A_h-1)
    right_coord = np.array([right_x, y])
    bottom_y = np.minimum(y + 1, A_w-1)
    bottom_coord = np.array([x, bottom_y])
    curr_coords = np.array([x,y])
    coords = np.array([curr_coords, right_coord, bottom_coord])
    self.findBestValues(coords)
  
  def doIteration(self, range_h, range_w, is_odd):
    for i in range_h:
      for j in range_w:
        if is_odd:
          self.oddPropagation(i, j)
        else:
          self.evenPropagation(i, j)
        self.randomSearch(i, j)
  
  def getPatchMatchFunction(self):
    (A_h, A_w) = self.A_size
    for itr in range(self.iterationsCount):
      print("iteration start: %d"%(itr + 1))
      if itr % 2 == 1:
        #iterations 1,3,5 (direct)
        range_h = range(A_h)
        range_w = range(A_w)
        self.doIteration(range_h, range_w, True)
      else:
        #iterations 2,4 (reverse)
        range_h = range(A_h - 1, -1, -1)
        range_w = range(A_w - 1, -1, -1)
        self.doIteration(range_h, range_w, False)
    return self.f
#END CLASS

def getSmallerImage(url):
  img = np.array(url_to_image(url))
  (h, w) = img.shape[:2]
  newShape = (int(h/4), int(w/4))
  return cv2.resize(img, newShape)


ITERATIONS = 2
PATCH_SIZE = 3
    
url_1 = "https://github.com/mdavydov/ComputerVisionCourse/blob/master/frames/v083.jpg?raw=true"
url_2 = "https://github.com/mdavydov/ComputerVisionCourse/blob/master/frames/v084.jpg?raw=true"
img_1 = getSmallerImage(url_1)
img_2 = getSmallerImage(url_2)
start = time.time()
pm = PatchMatch(img_1.copy(), img_2.copy(), ITERATIONS, PATCH_SIZE)
f = pm.getPatchMatchFunction()
end = time.time()
print(end - start)
```

```python
def reconstructImage(image, B, f):
  (A_h, A_w) = image.shape[:2]
  for i in range(A_h):
    for j in range(A_w):
      x_value = f[i, j][0]
      y_value = f[i, j][1]
      image[i, j, :] = B[x_value, y_value, :]

res_img = np.zeros_like(img_1)
#To test if this algorithm works, I used two frames from test folder (frames 83 and 84)
reconstructImage(res_img, img_2, f)
cv2_imshow(res_img)
#Result of trying to reconstruct frame:
```
